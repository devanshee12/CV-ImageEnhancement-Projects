```python
# Install required libraries
!pip install deepface
!pip install opencv-python-headless

# Import libraries
import cv2
from deepface import DeepFace
from google.colab.patches import cv2_imshow

# Upload image
from google.colab import files
uploaded = files.upload()
photo_path = list(uploaded.keys())[0]

# Analyze emotion
result = DeepFace.analyze(img_path=photo_path, actions=['emotion'])
dominant_emotion = result[0]['dominant_emotion']

# Draw the result on the image
img = cv2.imread(photo_path)
cv2.putText(img, f"Emotion: {dominant_emotion}", (50, 50),
            cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

# Show result
cv2_imshow(img)
print("Detected Emotion:", dominant_emotion)
```
